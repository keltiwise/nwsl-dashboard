```python
import requests
import pandas as pd

# Define the base URL for the API
base_url = "https://app.americansocceranalysis.com/api/v1/nwsl/games/shots"

# Send a GET request to the API
response = requests.get(base_url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()
    # If the data is a list of dictionaries (rows), we can directly convert it to a DataFrame
    # If it's nested, adjust accordingly based on the structure
    if isinstance(data, list):  # Case if the response is a list of rows
        df = pd.DataFrame(data)
    else:
        # If it's not a list but some other structure, check the nested data and flatten it
        print("Unexpected data format:", type(data))
        # This part can be adjusted based on the actual structure of the API response
        df = pd.DataFrame(data.get('results', []))  # For example, if it's wrapped under 'results'
    # Print the first few rows to verify
    print(df.head())
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")

# Define the filename for the CSV file
file_name = "shots_data_2024.csv"

# Save the DataFrame to a CSV file
df.to_csv("/Users/keltiwise/Desktop/NWSL/shots_data_2024.csv", index=False)

print(f"Data has been saved to {"/Users/keltiwise/Desktop/NWSL/shots_data_2024.csv"}")

```
